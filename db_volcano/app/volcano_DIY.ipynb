## 4. Create trace(s) for your graph 
In Plotly **trace** is a data series plotted as a particular type of graph. In this example we will use **scatter** but there are many more predefined [trace types](https://plotly.com/python/reference/scatter/) in *Plotly*. <br>
*<span style="color:gray">In general, complex graphs composed of many traces are more convenient to create using Graph Objects wrapper. You can find a suitable example in this repository.</span>*<br>
In this example, we will create an interactive scatter plot with only one trace, but the data will be split into smaller series and colored accordingly.

```python
fig = db.VolcanoPlot(              # define Figure object composed of a single volcano trace
    dataframe=df,                  # provide source dataFrame
    effect_size=x_data,            # a string denoting the column name for the effect size plotted on the x-axis
    p=y_data,                      # a string denoting the column name for the float to be plotted on the y-axis
    snp=snp_ids,                   # a string denoting the column name hat identifies each point being plotted
    gene=gene_ids,                 # a string denoting the column name for the GENE names
    logp=use_logp,                 # if True, the -log10 of the p-value is plotted
    point_size=marker_size,        # size of the points of the Scatter plot
    col=marker_color,              # color of the points of the volcano plot
    effect_size_line=significant_range,               # upper and lower bounds of the effect size values
    effect_size_line_width=significant_line_width,    # width of the effect size lines
    effect_size_line_color=significant_line_color,    # color of the effect size lines
    genomewideline_value=genomewideline_value,        # p-value above which the data points are considered significant
    genomewideline_width=genomewideline_width,        # width of the genome-wide line
    genomewideline_color=genomewideline_color,        # color of the genome-wide line
    highlight=highlight,                              # highlights the data points considered significant
    highlight_color=highlight_color,                  # highlight color of the data points considered significant
    annotation=annotations,                           # a string denoting the column to use as annotations
)
```

## 4. Update config for built-in options for the interactivity
At the top right of the interactive chart, when you hover over it, a manu appears with pre-built options for interactivity. By defining **config**, you can add or exclude certain options and make the chart even more customizable. In particular, *`modeBarButtonsToAdd`* allows you to add additional buttons with options that, for example, allow you to draw lines or shapes on the graph. This way you can highlight important points  or parts of the chart.

```python
config = {
    'responsive': True, 'showTips': True,
    'modeBarButtonsToAdd':['drawline', 'drawopenpath', 'drawclosedpath', 'drawcircle', 'drawrect',
                           'eraseshape', 'resetViews', 'toggleHover', 'toggleSpikelines'], 
    'toImageButtonOptions': {
        'format': static_img_format,        # one of png, svg, jpeg, webp
        'filename': static_img_filename,    # enter the default name for static image available for download
        'height': static_img_height,        # dimensions of the exported static image can be different than for the chart available in this notebook
        'width': static_img_width,          
        'scale': static_img_scale,          # multiply font size for the title/legend/axis/canvas sizes by this factor
    }
}
```
